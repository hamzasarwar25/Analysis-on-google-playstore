```python
android_clean=[]
already_added=[]

for index,row in android_df.iterrows():
    name=row["App"]
    n_reviews=row["Reviews"]
    
    if (review_max[name]==n_reviews) and (name not in already_added):
        android_clean.append(row)
        already_added.append(name)

```

```python
#For English app
def is_english(app_name):
    lst=[]
    for i in app_name:
        if ord(i)>127:
            lst.append(False)
        else:
            lst.append(True)
            
    non_ascii=0
    for j in lst:
        if j==False:
            non_ascii += 1
                
    if non_ascii>3:
        return False
    else:
            
        return True
```
